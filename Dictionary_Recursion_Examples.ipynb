## Recursion


Example 1 : multiply with while loop & recursion
<br>
a * b = a + a + a + a + a + a ... (b times)
 

```python
def mult_iter(a, b):
  result = 0
  while b > 0:
    result += a
    b -= 1
  return result

def mult_recur(a, b):
  if b == 1:
    return a
  else:
    return a + mult_recur(a, b-1) 
```

```python
print('Loop method of 2 x 10 =',mult_iter(2, 10))
print('recursion method of 2 x 10 =',mult_recur(2, 10))
```

### Example 2: fibonacci sequence

$$f(n)
\begin{cases}
1\quad \quad \quad \quad \quad \quad \quad  \quad if\ n\ =\ 1\  or\  2\\
f(n-1)+f(n-2)\quad \  if\ n\ \geq\ 3 \\
\end{cases}
$$


```python
def fib(num):
    '''
    Returns nth number of fibonacci sequence
    
    Parameter:
        num: nth of fibonacci sequence (type:int)
    '''
    if num ==1 or num == 2:
        return 1
    else:
        return fib(num-1) + fib(num-2)
```

```python
print('Show 1st of fibonacci list', fib(1))
print('Show 20th of fibonacci list', fib(20))
```
